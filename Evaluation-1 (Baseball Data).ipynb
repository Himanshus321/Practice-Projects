```python
# from above heatmap showing that some features are highly correlate and some are negatively correlate.

x=df.drop(['RA','ER','ERA'],axis=1)
y=df['W']
```

# Train Test Spilt

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=421)
```

```python
x_train.shape # Checking shape of training data.
```

```python
x_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score, accuracy_score
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
# adjusted R2 Score
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test) # Check how well our model fits on Test Data
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
plt.scatter(y_test,y_pred)
```

```python
# checking score by Lasso and ridge

from sklearn.linear_model import Lasso,Ridge, RidgeCV, LassoCV
```

```python
lassCV=LassoCV(alphas=None,max_iter=100,normalize=True)
lassCV.fit(x_train,y_train)
```

```python
alpha=lassCV.alpha_
alpha
```

```python
lasso_reg=Lasso(alpha)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
ridgecv=RidgeCV(alphas=np.arange(0.001,0.1,0.01),normalize=True)
ridgecv.fit(x_train,y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_model=Ridge(alpha=ridgecv.alpha_)
ridge_model.fit(x_train,y_train)
```

```python
ridge_model.score(x_test,y_test)
```
